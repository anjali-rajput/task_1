```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('E:\Powerbi/data.csv',error_bad_lines=False)
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(data['strength'])
```

```python
password_tuple = np.array(data)
```

```python
password_tuple
```

```python
import random
random.shuffle(password_tuple)
```

```python
X=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
   
```

```python
X
```

```python
y
```

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('megzy123')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer= word_divide_char)
```

```python
Z=vectorizer.fit_transform(X)
```

```python
Z.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=Z[0]
first_document_vector
```

```python
first_document_vector.T.todense()       #calling column data..where T is used for transpose and todense is a func
```

```python
df= pd.DataFrame(first_document_vector.T.todense(), index= vectorizer.get_feature_names(),columns = ['Tf*IDF'] )
```

```python
df.sort_values(by=['Tf*IDF'],ascending= False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(Z,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
dt=np.array(['Anji@907'])
pred = vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred= clf.predict(X_test)
y_pred
```

```python
#performance of model
#confusion matrix;accuracy score;classification repot
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm= confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
